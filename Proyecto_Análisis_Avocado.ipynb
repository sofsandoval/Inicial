```python
aguacates=pd.read_csv('aguacates.csv')
display(aguacates)
```

```python
import pandas as pd
from matplotlib import pyplot as plt
#Â¿En quÃ© regiÃ³n se venden mÃ¡s aguacates?


Vendidos_porRegion = (aguacates.groupby(["region"])[["total_bags_sold"]]
     .sum()
     .sort_values(by = "total_bags_sold"))


Vendidos_porRegion.plot(kind='barh', legend='Reverse')
plt.title(u'Ventas por region', size = 12.0)
plt.xlabel(u'Suma', size = 12.0)
plt.ylabel('Region', size = 12.0)
fig_size = plt.rcParams["figure.figsize"]
print ("Current size:", fig_size)
fig_size[0] = 20.0
fig_size[1] = 25.0
plt.savefig('detalle', bbox_inches='tight')
```

```python
#Â¿En quÃ© aÃ±o se venden mÃ¡s aguacates?


Vendidos_porAÃ±o = (aguacates.groupby(["year"])[["total_bags_sold"]]
     .sum()
     .sort_values(by = "total_bags_sold"))
Vendidos_porAÃ±o

Vendidos_porAÃ±o.plot(kind='barh', legend='Reverse')
plt.title(u'Ventas por aÃ±o', size = 12.0)
plt.xlabel(u'Suma', size = 12.0)
plt.ylabel('Region', size = 12.0)
fig_size = plt.rcParams["figure.figsize"]
print ("Current size:", fig_size)
fig_size[0] = 5.0
fig_size[1] = 5.0
plt.savefig('detalle', bbox_inches='tight')
```

```python
import pandas as pd
#extraemos los datos de nuestro
df=pd.read_csv('aguacates.csv')
#definimos nuestras variables 
sy=df['avg_price'].sum()
sx=df['count'].sum()
df['y2']=df['avg_price'].apply(lambda x: pow(x,2))
df['x2']=df['count'].apply(lambda x: pow(x,2))
s2x=df['x2'].sum()
s2y=df['y2'].sum()
df['xy']=df['count']*df['avg_price']
sxy=df['xy'].sum()
n=len(df['count'])
xp=(df['count'].mean())
yp=(df['avg_price'].mean())

b1=(sxy-(1/n)*(sx*sy))/((s2x)-(1/n)*pow(sx,2))
b0=yp-b1*xp
print("b1: ", b1)
print("b0: ",  b0)
```

```python
from numpy.polynomial.polynomial import polyfit
import matplotlib.pyplot as plt

x=df['count']
y=df['avg_price']
b,m = polyfit(x,y,1)
plt.plot(x,y,'.')
plt.plot(x,b+m*x,'-')
plt.show()
```
